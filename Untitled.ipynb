```python
from datetime import date
from pymongo import MongoClient
from morning.back_data.holidays import is_holidays, get_yesterday
from morning.config import db
from morning.back_data.fetch_stock_data import get_day_period_data, get_day_minute_period_data
import pandas as pd


remote_codes = []
stock = MongoClient(db.HOME_MONGO_ADDRESS)['stock']
codes = list(stock['KOSDAQ_CODES'].find())
for code in codes:
    remote_codes.append(code['code'])

from_date = date(2018, 1, 1)
current = from_date
df = pd.DataFrame(columns=['date', 'code', 'profit', 'serial_days', 'amount_avg', 'prevday_amount', 'today_amount'])

while from_date < date(2019, 12, 8):
    if is_holidays(from_date):
        from_date += timedelta(days = 1)
        continue
    
    if from_date.month != current.month:
        df.to_excel(current.strftime('%Y%m') + '_inst.xlsx')
        current = from_date
        df.drop(df.index, inplace=True)
        
    kosdaq_overall = get_day_period_data('U201', from_date - timedelta(days=365), get_yesterday(from_date))
    kosdaq_overall = sorted(kosdaq_overall, key = lambda i: i['0']) 
    kosdaq_overall_df = pd.DataFrame(kosdaq_overall)
    kosdaq_moving_average = kosdaq_overall_df['5'].rolling(20).mean()
    moving_average = kosdaq_moving_average.iloc[-1]
    yesterday_kosdaq = kosdaq_overall_df['5'].iloc[-1]


    
    for code in remote_codes:
        data = get_day_period_data(code, from_date - timedelta(days=20), from_date)
        if len(data) < 6:
            continue
        data = sorted(data, key = lambda i: i['0'], reverse=True) 
        today_profit = (data[0]['5'] - data[0]['2']) / data[0]['2'] * 100.
        today_profit = float("{0:.2f}".format(today_profit))
        serial_days = 0
        amount_avg = 0
        for d in data[-2::-1]:
            if d['12'] > 0:
                serial_days += 1
                amount_avg += d['12'] * d['5']
            else:
                break
        if serial_days > 0:
            amount_avg = amount_avg / serial_days
            amount_avg = float("{0:.2f}".format(amount_avg))
        else:
            amount_avg = 0
        
        summary = {'date': from_date, 'code': code, 'profit': today_profit, 'serial_days': serial_days, 'amount_avg': amount_avg, 'prevday_amount': data[-2]['7'], 'today_amount': data[0]['7'], 'prevday_kosdaq': yesterday_kosdaq, 'kosdaq_20ma': moving_average}
        df = df.append(summary, ignore_index=True)
        
    
    from_date += timedelta(days = 1)

```
