```python
from datetime import date
from pymongo import MongoClient
from morning.back_data.holidays import is_holidays
from morning.config import db
from morning.back_data.fetch_stock_data import get_day_period_data, get_day_minute_period_data

remote_codes = []
stock = MongoClient(db.HOME_MONGO_ADDRESS)['stock']
codes = list(stock['KOSDAQ_CODES'].find())
for code in codes:
    remote_codes.append(code['code'])

for code in remote_codes:
    print(code)
    get_day_period_data(code, date(2018, 1, 1), date(2019, 12, 7))
    #get_day_minute_period_data(code, date(2018, 1, 1), date(2019, 12, 7))
```

```python
from_date = date(2019, 5, 22)

while from_date < date(2019, 12, 7):
    if is_holidays(from_date):
        from_date += timedelta(days = 1)
        continue
    
    datas = []
    for code in remote_codes:
        data = get_day_period_data(code, from_date, from_date)
        if len(data) > 0:
            data[0]['code'] = code
            datas.append(data[0])

    datas = sorted(datas, key=lambda i: i['7'], reverse=True)
    bull_codes = [d['code'] for d in datas]
    bull_codes = bull_codes[:40]
    
    for b in bull_codes:
        print(from_date, b)
        get_day_minute_period_data(b, from_date, from_date)
    
    from_date += timedelta(days = 1)
```
