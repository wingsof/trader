```python
def get_profit_summary(df, year):
    profit_summary = {}
    df = df[(trade_df['buy_date'] >= datetime(year, 1, 1)) & (trade_df['buy_date'] <= datetime(year+1, 1, 1))]
    for i, v in df['profit'].items():
        profit_list = eval(v)
        for i, profit in enumerate(profit_list):
            if i in profit_summary:
                profit_summary[i].append(profit)
            else:
                profit_summary[i] = [profit]

    for k, v in profit_summary.items():
        print(k, sum(v) / len(v), len(v))
    return profit_summary
```

```python
def check_trade_count_per_day(df):
    from_date = date(2018, 1, 2)
    until_date = date(2020,1, 7)
    trade_count = []
    while from_date <= until_date:
        trading_day_len = len(df[(df['buy_date'] <= from_date) & (df['sell_date'] >= from_date)])
        trade_count.append(trading_day_len)
        from_date += timedelta(days=1)
    return np.array(trade_count).mean()
```

```python
def worst_analysis(df, year):
    df = df[(trade_df['buy_date'] >= datetime(year, 1, 1)) & (trade_df['buy_date'] <= datetime(year+1, 1, 1))]
    list_by_profit = []
    for i, v in df['profit'].items():
        profit_list = eval(v)
        list_by_profit.append({'mean': np.array(profit_list).mean(), 'profit': profit_list})
    list_by_profit = sorted(list_by_profit, key=lambda x: x['mean'], reverse=True)
    return list_by_profit
```

```python
from gevent import monkey; monkey.patch_all()
from morning_server import message
from morning_server import stock_api
from morning_server import stream_readwriter
from morning.back_data import holidays
from morning.pipeline.converter import dt
from utils import time_converter
import socket
from morning_server import trendfinder
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
def show_graph(code, buy_date, sell_date):
    sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)

    server_address = (message.SERVER_IP, message.CLIENT_SOCKET_PORT)
    sock.connect(server_address)

    message_reader = stream_readwriter.MessageReader(sock)
    message_reader.start()
    data = stock_api.request_stock_day_data(message_reader, code, buy_date - timedelta(days=20), sell_date + timedelta(days=20))
    data_c = []
    buy_price, sell_price = 0, 0
    for d in data:
        converted = dt.cybos_stock_day_tick_convert(d)
        converted['date'] = time_converter.intdate_to_datetime(converted['0']).date()
        if converted['date'] == buy_date:
            buy_price = converted['close_price']
        elif converted['date'] == sell_date:
            sell_price = converted['close_price']
        data_c.append(converted)
        
    fig = make_subplots(rows=3, cols=1)
    fig.add_trace(go.Candlestick(x=[d['date'] for d in data_c], open=[d['start_price'] for d in data_c],
                                low=[d['lowest_price'] for d in data_c], high=[d['highest_price'] for d in data_c],
                                close=[d['close_price'] for d in data_c], increasing_line_color='red', decreasing_line_color='blue'), row=1, col=1)
    fig.update_xaxes(type='category', row=1, col=1)
    fig.add_trace(go.Bar(x=[d['date'] for d in data_c], y=[d['volume'] for d in data_c]), row=3, col=1)
    fig.update_xaxes(type='category', row=3, col=1)
    annotations = [dict(x=buy_date, y=buy_price, xref='x', yref='y', showarrow=True, xanchor='left', text='buy'),
                  dict(x=sell_date, y=sell_price, xref='x', yref='y', showarrow=True, xanchor='left', text='sell')]
    fig.update_layout(title=code, yaxis_tickformat='d', annotations=annotations)
    return fig
```

```python
show_graph('A005290', date(2019, 3, 22), date(2019, 3, 27))
```
