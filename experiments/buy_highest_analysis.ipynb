```python
def get_profit_summary(df, year):
    profit_summary = {}
    df = df[(trade_df['buy_date'] >= datetime(year, 1, 1)) & (trade_df['buy_date'] <= datetime(year+1, 1, 1))]
    for i, v in df['profit'].items():
        profit_list = eval(v)
        for i, profit in enumerate(profit_list):
            if i in profit_summary:
                profit_summary[i].append(profit)
            else:
                profit_summary[i] = [profit]

    for k, v in profit_summary.items():
        print(k, sum(v) / len(v), len(v))
    return profit_summary
```
